```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
df = pd.read_csv(url,'\t')
df.head (4)
```

## 1. Filter 1 name in "item_name" and see the item_price

#### LOC AND ILOC only works with the INDEX 

```python
df.loc[df['order_id'] == 3]
```

```python
df.loc[(df['order_id'] == 3 ), ['item_name', 'item_price']] # SPECIFIC 
```

```python
df.loc[(df['item_name'] == 'Chips' ) & (df['quantity'] == 1 ) , ['item_name','item_price']] #SPECIFIC with two conditions 
#Another way -- df.loc[df.item_name.isin(['Chips']),['item_name','item_price']]
```

```python
df[df['item_name'].isin(['Izze'])] #SPECIFIC search in a specific column IS IN 
```

```python
df[~df['item_name'].isin(['Izze'])] #show the list opposite of above 
```

```python
df[df['choice_description'].str.contains('App',na=False)] #search for any item or letter in a column you want in a ROW with NAN values. 
```

```python
df[df['item_name'].str.contains('Chips')] #search for any item or letter in a column you want in a ROW
```

```python
df[df['item_name'].str.contains('Chips')][['choice_description','item_price']] #returns in the required columns
```

```python
df.filter(like='price') #search for any item or letter in the COLUMNS 
```

```python
df2 = df[df.apply(lambda row: row.astype(str).str.contains('Chips').any(), axis=1)] #search for any that contains CHIPS in any row
df2 
```
